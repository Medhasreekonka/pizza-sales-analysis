```python
KPI’S REQUIREMENT: We need to analyze key indicators for our pizza sales data to gain insights into our businesss performance.Specifically ,WE want to calculate the following metrics.

TOTAL REVENUE: The sum of the total price of all pizza orders.
AVERAGE ORDER VALUE: The average amount spent per order, calculated by dividing the total revenue by the total number of orders.
TOTAL PIZZA’S SOLD: The sum of the quantities of all pizas sold
TOTAL ORDER”S: The total number of orders placed
AVERAGE PIZZAS PER ORDER: The average number of pizzas sold per order,calculated by dividing the total number of pizzas sold by total number of orders.

CHARTS REQUIREMENT: WE Would like to visualize various aspects of our pizza sales data to gain insights and understand key trends. We have identified the following requirements for creating charts:
1.DAILY TREND FOR TOTAL ORDERS:

Create a bar chart that displays the daily trend of total orders over a specific time period. This chart will help us identity any patterns or fluctuations in order volumes on a daily basis

2.MONTHLY TREND FOR TOTAL ORDERS:

Create a line chart that illustrates the hourly trend of total orders throughout the day.This chart will allow you to identify peak hours or periods of high order activity.

3.PERCENTAGE OF SALES BY PIZZA CATEGORY:

Create a pie chart that shows the distribution of sales across different pizza categories. This chart will provide insights into the popularity of various pizza categories and their contribution to overall sales.

4.PERCENTAGE OF SALES BY PIZZA SIZE:

Generate a pie chart that represents the percentage of sales attributed to different pizza sizes.This chart will help us to understand customer preference for pizza sizes and their impact on sales.

5.TOTAL PIZZA SOLD BY PIZZA CATEGORY:
Create a funnel chart that presents the total number of pizzas sold for each pizza category. This Chart will allow you to compare the sales performance of different pizza categories

6.TOP 5 BEST SELLERS BY REVENUE,TOTAL QUANTITY AND TOTAL ORDERS:

Create A Bar chart highlighting the top 5 best selling pizzas based on the revenue,total quantity ,total orders. This chart will allows you to identify the most popularpizza options

7.BOTTOM 5 BEST SELLERS BY REVENUE,TOTAL QUANTITY AND TOTAL ORDERS:

Create a bar chart showcasing the bottom 5 worst selling pizzas based on the revenue total quantity,total orders.this chart will enable us to identify underperforming or less popular pizza options.
```

```python
# KPI Calculations
total_revenue = df['total_price'].sum()
total_orders = df['order_id'].nunique()
total_pizzas_sold = df['quantity'].sum()
average_order_value = total_revenue / total_orders
average_pizzas_per_order = total_pizzas_sold / total_orders
```

```python
# Store KPI results
kpi_results = {
    "Total Revenue": total_revenue,
    "Total Orders": total_orders,
    "Total Pizzas Sold": total_pizzas_sold,
    "Average Order Value": average_order_value,
    "Average Pizzas per Order": average_pizzas_per_order,
}
```

```python
# Convert order_date column to datetime format
df['order_date'] = pd.to_datetime(df['order_date'], format='%d-%m-%Y')

# Extract day name (Monday, Tuesday, etc.) from the date
df['day'] = df['order_date'].dt.day_name()

# Group data by day and count unique orders
daily_orders = df.groupby('day')['order_id'].nunique()

# Create bar chart
plt.figure(figsize=(8,5))
daily_orders.plot(kind='bar')

# Chart labels
plt.title("Daily Trend for Total Orders")
plt.xlabel("Day of Week")
plt.ylabel("Total Orders")

# Display chart
plt.show()
```

```python
df['month_short'] = df['order_date'].dt.strftime('%b')

monthly_orders = df.groupby(['month_number','month_short'])['order_id'].nunique().reset_index()
monthly_orders = monthly_orders.sort_values('month_number')

plt.figure(figsize=(10,5))
plt.plot(monthly_orders['month_short'], monthly_orders['order_id'], marker='o')

plt.title("Monthly Trend for Total Orders")
plt.xlabel("Month")
plt.ylabel("Total Orders")

plt.show()
```

```python
# Calculate total sales revenue for each pizza category
category_sales = df.groupby('pizza_category')['total_price'].sum()

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(category_sales,
        labels=category_sales.index,
        autopct='%1.1f%%')  # Show percentage

# Chart title
plt.title("Percentage of Sales by Pizza Category")

# Display chart
plt.show()
```

```python
# Calculate total revenue for each pizza size
size_sales = df.groupby('pizza_size')['total_price'].sum()

# Create pie chart
plt.figure(figsize=(6,6))
plt.pie(size_sales,
        labels=size_sales.index,
        autopct='%1.1f%%')

# Chart title
plt.title("Percentage of Sales by Pizza Size")

# Display chart
plt.show()
```

```python
# Import plotly
import plotly.express as px

# Calculate total pizzas sold for each category
category_quantity = df.groupby('pizza_category')['quantity'].sum().reset_index()

# Sort values so the funnel looks correct (largest at top)
category_quantity = category_quantity.sort_values(by='quantity', ascending=False)

# Create funnel chart
fig = px.funnel(
    category_quantity,
    x='quantity',           # Quantity sold
    y='pizza_category',     # Pizza categories
    title="Total Pizzas Sold by Pizza Category"
)

# Display the chart
fig.show()
```

```python
# Calculate revenue for each pizza
top_revenue = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(5)

# Create bar chart
plt.figure(figsize=(8,5))
top_revenue.plot(kind='bar')

plt.title("Top 5 Best Selling Pizzas by Revenue")
plt.xlabel("Pizza Name")
plt.ylabel("Revenue")

plt.show()
```

```python
# Calculate total quantity sold for each pizza
top_quantity = df.groupby('pizza_name')['quantity'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_quantity.plot(kind='bar')

plt.title("Top 5 Best Selling Pizzas by Quantity")
plt.xlabel("Pizza Name")
plt.ylabel("Quantity Sold")

plt.show()
```

```python
# Count total orders for each pizza
top_orders = df.groupby('pizza_name')['order_id'].count().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,5))
top_orders.plot(kind='bar')

plt.title("Top 5 Best Selling Pizzas by Orders")
plt.xlabel("Pizza Name")
plt.ylabel("Total Orders")

plt.show()
```

```python
# Find pizzas with lowest revenue
bottom_revenue = df.groupby('pizza_name')['total_price'].sum().sort_values().head(5)

plt.figure(figsize=(8,5))
bottom_revenue.plot(kind='bar')

plt.title("Bottom 5 Pizzas by Revenue")
plt.xlabel("Pizza Name")
plt.ylabel("Revenue")

plt.show()
```

```python
# Find pizzas with lowest quantity sold
bottom_quantity = df.groupby('pizza_name')['quantity'].sum().sort_values().head(5)

plt.figure(figsize=(8,5))
bottom_quantity.plot(kind='bar')

plt.title("Bottom 5 Pizzas by Quantity Sold")
plt.xlabel("Pizza Name")
plt.ylabel("Quantity Sold")

plt.show()
```

```python
# Find pizzas with lowest number of orders
bottom_orders = df.groupby('pizza_name')['order_id'].count().sort_values().head(5)

plt.figure(figsize=(8,5))
bottom_orders.plot(kind='bar')

plt.title("Bottom 5 Pizzas by Orders")
plt.xlabel("Pizza Name")
plt.ylabel("Total Orders")

plt.show()
```
